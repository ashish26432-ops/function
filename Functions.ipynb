# Question 5:       Given: sales = [12000, 15000, 11000, 18000, 17000] Using a loop:

```python
# 1. Calculate total revenue.
sales = [12000, 15000, 11000, 18000, 17000]
total_revenue=0
for sale in sales:
  total_revenue+=sale
print(total_revenue)
# 2. Count how many months had sales above 15000
count=0
for sale in sales:
  if sale >15000:
    count+=1
print("list of months had sales above 15000=",count)

# 3. Find maximum sales without using max()
maximum=sales[0]
for sale in sales:
  if sale>maximum:
    maximum=sale
print("maximum sales=",maximum)
```

# Question 8: Given
prices = [500, 1500, 2500, 800, 1200]

```python
#1. Use map() + lambda to add 18% GST to all prices.
prices = [500, 1500, 2500, 800, 1200]
gst_prices = list(map(lambda x: x * 1.18, prices))
print(gst_prices)

#2. Use filter() to select prices above 1000
filtered_prices = list(filter(lambda x: x > 1000, prices))
print(filtered_prices)
```

# Question 9: Write a generator that yields squares of numbers from 1 to 1,000,000.


```python
#1. Write a generator that yields squares of numbers from 1 to 1,000,000
def square_generator():
    for i in range(1, 1000001):
        yield i * i

# Example: Print first 5 squares
gen = square_generator()

for _ in range(5):
    print(next(gen))
```

#Question 10: Design a class: class Product: With: ● Attributes: name, price, quantity ● Method: calculate_revenue() Create 3 objects and print their revenue


```python
class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def calculate_revenue(self):
        return self.price * self.quantity



product1 = Product("Laptop", 50000, 2)
product2 = Product("Mobile", 20000, 5)
product3 = Product("Headphones", 1500, 10)

print(product1.name, "Revenue =", product1.calculate_revenue())
print(product2.name, "Revenue =", product2.calculate_revenue())
print(product3.name, "Revenue =", product3.calculate_revenue())
```
